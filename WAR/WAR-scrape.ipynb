```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Firefox()
driver.get("https://www.baseball-reference.com/leagues/majors/2022-value-batting.shtml")

html = driver.page_source
soup = BeautifulSoup(html, "lxml")

pbp_table = soup.find_all("table", {"id":"players_value_batting"})
```

```python
war_list = soup.find_all('tr', {"class":"full_table"})
```

```python
rows = []

for tr in war_list:
    rows.append(tr)
```

```python
rows = list(zip(*rows[1:])) # tranpose the table
df = pd.DataFrame(rows[1:], columns=rows[0])
df
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df
```
